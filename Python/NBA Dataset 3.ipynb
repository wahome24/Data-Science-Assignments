### Players Analysis

```python
#Importing Libraries

import pandas as pd
```

```python
#Loading in the data

df = pd.read_csv(r"C:\Users\USER\Desktop\SD DATA\NBA\DataSet 3.csv")

df
```

```python
#Selecting the required columns

new_df = df[['season','player','experience','tm','Games','mp','fg','fga','pts']]

new_df
```

```python
new_df.columns
```

```python
#Updating the column names

new_df.columns = ['Season','Player','Experience','Team','Games_Played','Minutes_Played','Field_Goals','Field_Goals_Attempted','Points']

new_df
```

### Filtering the data

#### Required Data : Team - New York Knicks, Season - 2021-2025

```python
season = [2021,2022,2023,2024,2025]
#Team abv - NYK
```

```python
new_df[(new_df['Team'] == 'NYK') & (new_df['Season'].isin(season))]
```

```python
#storing the cleaned data into a new dataframe

cleaned_df = new_df[(new_df['Team'] == 'NYK') & (new_df['Season'].isin(season))]

cleaned_df
```

```python
#Resetting the Index

cleaned_df = cleaned_df.reset_index(drop=True)

cleaned_df
```

```python
#Top 10 players with the highest games played

players = pd.DataFrame(cleaned_df.groupby('Player')['Games_Played'].sum())  

results = players.sort_values('Games_Played', ascending=False)

results.head(10)
```

```python
#Team's total points over the seasons

cleaned_df.groupby('Season')['Points'].sum()
```

```python
#Players Games played vs field goals vs points

cleaned_df.groupby('Player')[['Games_Played','Field_Goals','Points']].sum()

players_detailed = pd.DataFrame(cleaned_df.groupby('Player')[['Games_Played','Field_Goals','Points']].sum())

players_detailed
```

```python
#Players sorted from highest to lowest points.

players_detailed.sort_values(['Points'], ascending=False)
```
