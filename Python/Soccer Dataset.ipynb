### Team Analysis

```python
#Importing Libraries

import pandas as pd
```

```python
#Loading in the data

df = pd.read_csv(r"C:\Users\USER\Desktop\SD DATA\SOCCER\Soccer Dataset.csv")

df
```

```python
#Selecting the required columns

new_df = df[['Date','Team','Team_Score','Team_Points','season','Location']]

new_df
```

### Filtering the data

#### Required Data : Team - Manchester United, Season -last five seasons (2018-2022)

```python
seasons = [2018,2019,2020,2021,2022]
```

```python
cleaned_df = new_df[(new_df['Team'] == 'MANCHESTER UTD') & (new_df['season'].isin(seasons))]

cleaned_df
```

```python
#Resetting the Index

cleaned_df = cleaned_df.reset_index(drop = True)

cleaned_df
```

```python
#Checking for blank fields

#cleaned_df['Date'].isna().sum()
#cleaned_df['Team'].isna().sum()
cleaned_df['Team_Score'].isna().sum() 
cleaned_df['Team_Points'].isna().sum() 
#cleaned_df['season'].isna().sum()
#cleaned_df['Location'].isna().sum()

#24 blank fields in team score and points
```

```python
#removing blanks

cleaned_df.dropna(inplace=True)

cleaned_df
```

```python
#Comparing Points $ goals Scored home vs away

cleaned_df.groupby('Location')[['Team_Score','Team_Points']].sum()
```

```python
#Comparing Points $ goals Scored over the seasons

cleaned_df.groupby('season')[['Team_Score','Team_Points']].sum()
```
